```python
import wfdb
import os
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import butter, filtfilt, find_peaks, iirnotch
from keras.layers import Conv1D, MaxPooling1D, Dense, Dropout, Flatten, BatchNormalization
from keras.models import Sequential
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
import seaborn as sns

mi_data = './data/mi'
hc_data = './data/healthy'

def load_ecg(data):
    all_leads_data = []
    global record
    patient_ids = os.listdir(data)
    for patient_id in patient_ids:
        path = os.path.join(data, patient_id)
        try:
            hea_files = [f for f in os.listdir(path) if f.endswith('.hea')]
            if hea_files:
                record_name = os.path.join(path, hea_files[0].replace('.hea', ''))
                record = wfdb.rdrecord(record_name)
                all_leads_signal = record.p_signal[:, :12]
                all_leads_data.append((patient_id, all_leads_signal))
                print(f"Loaded all 12 leads for patient {patient_id}: {record_name}")
            else:
                print(f"No .hea file found for patient {patient_id} in directory {path}")
        except:
            print(f"Failed to load all 12 leads for patient {patient_id}")
    return all_leads_data

print("Loading MI patients' all 12 lead data...")
mi_all_leads_data = load_ecg(mi_data)

print("Loading healthy patients' all 12 lead data...")
hc_all_leads_data = load_ecg(hc_data)

print(f"Loaded all 12 lead data for {len(mi_all_leads_data)} MI patients.")
print(f"Loaded all 12 lead data for {len(hc_all_leads_data)} healthy patients.")

def norm_ecg(all_leads_data):
    norm_data = []
    for patient_id, all_leads_signal in all_leads_data:
        norm_leads = (all_leads_signal - np.min(all_leads_signal)) / (np.max(all_leads_signal) - np.min(all_leads_signal))
        norm_data.append((patient_id, norm_leads))
    return norm_data

norm_data_mi = norm_ecg(mi_all_leads_data)
norm_data_hc = norm_ecg(hc_all_leads_data)

fs = record.fs

def highpass_filter(data, cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='high', analog=False)
    y = filtfilt(b, a, data, axis=0)
    return y

def notch_filter(data, cutoff, fs, Q=50):
    b_notch, a_notch = iirnotch(cutoff, Q, fs)
    y = filtfilt(b_notch, a_notch, data, axis=0)
    return y

def filter_ecg(all_leads_data):
    filtered_data = []
    for patient_id, all_leads_signal in all_leads_data:
        filtered_ecg = highpass_filter(all_leads_signal, 0.5, fs, 5)
        filtered_ecg = notch_filter(filtered_ecg, 50, fs, 50)
        filtered_data.append((patient_id, filtered_ecg))
    return filtered_data

filtered_data_mi = filter_ecg(norm_data_mi)
filtered_data_hc = filter_ecg(norm_data_hc)

def rpeaks_detect(signal):
    rpeaks, _ = find_peaks(signal[:, 0], distance=fs * 0.5)
    return rpeaks

def segmentation(rpeaks, signal, window_size):
    segments = []
    for rpeak in rpeaks:
        start = rpeak - 200
        end = rpeak + 451
        if start >= 0 and end <= len(signal):
            segment = signal[start:end, :]
            segments.append(segment)
    return segments

global window_size
window_size=651

def beat_seg(all_leads_data):
    beat_segments = []
    for patient_id, signal in all_leads_data:
        r_peaks = rpeaks_detect(signal)
        beat_seg = segmentation(r_peaks, signal, window_size)
        global x
        x=signal.shape[1]
        for lead_index in range(x):
            for segment in beat_seg:
                beat_segments.append((patient_id, lead_index, segment[:, lead_index]))
    return np.array(beat_segments, dtype=object)

beat_seg_mi = beat_seg(filtered_data_mi)
beat_seg_hc = beat_seg(filtered_data_hc)

def norm_beat(beat_seg):
    norm_beats = []
    for patient_id, lead_index, segment in beat_seg:
        mean_beat = np.mean(segment)
        std_beat = np.std(segment)
        norml_beat = (segment - mean_beat) / std_beat
        norm_beats.append((patient_id, lead_index, norml_beat))
    return np.array(norm_beats, dtype=object)

norm_beat_mi = norm_beat(beat_seg_mi)
norm_beat_hc = norm_beat(beat_seg_hc)

def reshape_beats(norm_beats):
    reshaped_beats = []
    for _, _, segment in norm_beats:
        reshaped_beats.append(segment)
    return np.array(reshaped_beats)

reshaped_norm_beat_mi = reshape_beats(norm_beat_mi).reshape(-1, window_size, x )
reshaped_norm_beat_hc = reshape_beats(norm_beat_hc).reshape(-1, window_size, x )


print("MI Patient Segments Shape:", reshaped_norm_beat_mi.shape)
print("HC Patient Segments Shape:", reshaped_norm_beat_hc.shape)

labels_mi = np.ones(reshaped_norm_beat_mi.shape[0])
labels_hc = np.zeros(reshaped_norm_beat_hc.shape[0])


X = np.concatenate((reshaped_norm_beat_mi, reshaped_norm_beat_hc), axis=0)
y = np.concatenate((labels_mi, labels_hc))


X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.15, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.1765, random_state=42) # 0.25 x 0.8 = 0.2


model = Sequential([
    Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])),
    MaxPooling1D(pool_size=2),
    
    Conv1D(filters=64, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    
    Conv1D(filters=128, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])


model.compile(optimizer=Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy'])

model.summary()

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, min_lr=0.00001)  



history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_val, y_val), callbacks=[early_stopping, reduce_lr])


test_loss, test_acc = model.evaluate(X_test, y_test, verbose=2)
print(f"Test accuracy: {test_acc}")


predictions = model.predict(X_test)

predicted_classes = (predictions > 0.5).astype("int32")

train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
train_loss = history.history['loss']
val_loss = history.history['val_loss']

print(f"Training Accuracy: {train_acc}")
print(f"Validation Accuracy: {val_acc}")
print(f"Training Loss: {train_loss}")
print(f"Validation Loss: {val_loss}")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(train_acc, label='Training Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')


plt.subplot(1, 2, 2)
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')

plt.show()


true_classes = y_test.astype("int32") 
print("Classification Report:")
print(classification_report(true_classes, predicted_classes))


conf_matrix = confusion_matrix(true_classes, predicted_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True,fmt="d", cmap="Blues",cbar_kws={"orientation": "vertical", "label":"color bar"}, xticklabels=['HC', 'MI'], yticklabels=['HC', 'MI'])
plt.title("Confusion Matrix")
plt.ylabel('Actual Classes')
plt.xlabel('Predicted Classes')
plt.show()
```
